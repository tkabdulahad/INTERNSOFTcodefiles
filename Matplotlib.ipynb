```python
normals = pd.Series(np.random.normal(size=25))
print(normals)
normals.plot()
```

```python
normals.cumsum().plot(grid=True)
```

```python
variables = pd.DataFrame({'normal': np.random.normal(size=15), 'gamma': np.random.gamma(1, size=15), 
                          'poisson': np.random.poisson(size=15)})
print(variables)

variables.cumsum(0).plot()
```

```python
variables.cumsum(0).plot(subplots=True)
```

```python
variables.cumsum(0).plot(secondary_y='normal')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(7, 12))
for i,var in enumerate(['normal','gamma','poisson']):
    variables[var].cumsum(0).plot(ax=axes[i], title=var)
axes[0].set_ylabel('cumulative sum')
```

```python
fig = plt.figure()
ax = plt.axes()
s = np.linspace(0, 30, 50)
print(s)
ax.plot(s, np.sin(s));
```

```python
plt.plot(s, np.sin(s - 0), linestyle='-')        
plt.plot(s, np.sin(s - 1), linestyle=':')           
plt.plot(s, np.sin(s - 3), linestyle='-.')  
```

```python
rng = np.random.RandomState(9)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5,cmap='Reds')
plt.colorbar()  
```

```python
x = np.linspace(0, 10, 100)
dy = 0.1
y = np.sin(x) + dy * np.random.randn(100)
plt.errorbar(x, y, yerr=dy, fmt='.b');
```
